```python
#Years of Olympics 
eduYear = [1980,1984,1988,1992,1996,2000,2004,2008,2012,1994,1998,2002,2006,2010,2014]
```

## Loading Education Index Data Data

```python
# Ensure the file exists
filepath = "../data/raw/UNEducationIndex.csv"
if not os.path.exists(filepath):
    print("Missing dataset file")
else:
    eduDF = pd.read_csv(filepath, encoding = "ISO-8859-1")
    print("File Read")
```

```python
eduDF.info()
```

```python
eduDF[['1980','1985','1990','1995','2000','2005','2006','2007','2008','2009','2010','2011','2012','2013']] = eduDF[['1980','1985','1990','1995','2000','2005','2006','2007','2008','2009','2010','2011','2012','2013']].apply(pd.to_numeric, errors='coerce')
eduDF.info()
```

```python
eduDF['Country'] = eduDF['Country'].astype(str)
eduDF.info()
```

```python
for year in eduYear:
    if str(year) in eduDF.columns[2:]:
        continue
    else:
        print(year)
```

```python
year1984 = []
for index,row in eduDF.iterrows():
    curr = row['1985']
    prev = row['1980']
    diff = (curr-prev)/5
    year = curr-diff
    year1984.append(year)
eduDF['1984'] = year1984
```

```python
year1988 = []
for index,row in eduDF.iterrows():
    curr = row['1990']
    prev = row['1985']
    diff = (curr-prev)/5
    year = curr-(diff*2)
    year1988.append(year)
eduDF['1988'] = year1988
```

```python
year1992 = []
year1994 = []
for index,row in eduDF.iterrows():
    curr = row['1995']
    prev = row['1990']
    diff = (curr-prev)/5
    year = curr-(diff*3)
    temp = curr-diff
    year1992.append(year)
    year1994.append(temp)
eduDF['1992'] = year1992
eduDF['1994'] = year1994
```

```python
year1996 = []
year1998 = []
for index,row in eduDF.iterrows():
    curr = row['2000']
    prev = row['1995']
    diff = (curr-prev)/5
    year = prev+diff
    temp = curr-(diff*2)
    year1996.append(year)
    year1998.append(temp)
eduDF['1996'] = year1996
eduDF['1998'] = year1998
```

```python
year2004 = []
year2002 = []
for index,row in eduDF.iterrows():
    curr = row['2005']
    prev = row['2000']
    diff = (curr-prev)/5
    year = prev+(diff*2)
    temp = curr-diff
    year2002.append(year)
    year2004.append(temp)
eduDF['2004'] = year2004
eduDF['2002'] = year2002
```

```python
cols = [ str(year) for year in eduYear]
cols = cols[:len(cols)-1]
cols = ['Country']+cols
```

```python
eduDF = eduDF[cols]
eduDF = eduDF.set_index('Country')
eduDF.head()
```

## Joining the DataFrames

```python
edu = []
for index,row in df.iterrows():
    if 'Korea, Rep.' in row['Country']:
        country = row['Country']
    else:
        country = row['Country'].split(",")[0]
    year = row['Year']
    try:
        index = eduDF.loc[country][str(year)]
    except:
        index = np.nan
    edu.append(index)
df['Education_Index']=edu
df.head(40)
```

```python
df.to_csv('Country/countries_250.csv', index=False)
```
