```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os.path
```

```python
filepath = '../../data/analysis/complete_data.csv'
```

# In a single games

```python
medalsAwarded = df.groupby("Host_City")['Gold','Total_Medals','Ath_Rating'].sum()
medalsAwarded.head()
```

```python
medalsWon = df.groupby(['Country',"Host_City"])['Gold','Total_Medals','Ath_Rating'].sum()
medalsWon.head()
```

```python
def getGoldshare(row):
    return  row.Gold/medalsAwarded.loc[row.Host_City ]['Gold']
```

```python
medalsWon = medalsWon.reset_index(level='Host_City')
goldshare = medalsWon.apply(getGoldshare,axis=1)
medalsWon['Gold_Share'] = goldshare
medalsWon.head()
```

```python
def getMedalshare(row):
    return  row.Total_Medals/medalsAwarded.loc[row.Host_City ]['Total_Medals']
```

```python
medalshare = medalsWon.apply(getMedalshare,axis=1)
medalsWon['Medal_Share'] = medalshare
medalsWon.head()
```

```python
def getRatingshare(row):
    return  row.Ath_Rating/medalsAwarded.loc[row.Host_City ]['Ath_Rating']
```

```python
ratingshare = medalsWon.apply(getRatingshare,axis=1)
medalsWon['Rating_Share'] = ratingshare
medalsWon.head()
```

```python
temp = medalsWon.sort_values('Gold_Share',ascending=False)[['Host_City','Gold_Share']].head(10)
temp['Gold_Share'] = temp['Gold_Share']*100
temp.set_index(['Host_City'], append=True).plot(kind='bar',color='Gold')
plt.show()
```

```python
temp = medalsWon.sort_values('Medal_Share',ascending=False)[['Host_City','Medal_Share']].head(10)
temp['Medal_Share'] = temp['Medal_Share']*100
temp.set_index(['Host_City'], append=True).plot(kind='bar',color='blue')
plt.show()
```

```python
temp = medalsWon.sort_values('Rating_Share',ascending=False)[['Host_City','Rating_Share']].head(10)
temp['Rating_Share'] = temp['Rating_Share']*100
temp.set_index(['Host_City'], append=True).plot(kind='bar',color='Grey')
plt.show()
```

```python
def standardBar(df,column,num):
    temp = df.sort_values(column,ascending=False).head(num)
    barlist = temp.column.plot(kind='bar',color='#878787')
    barlist.get_children()[0].set_color('Gold')
    barlist.get_children()[0].set_color('#c0c0c0')
    barlist.get_children()[0].set_color('#cd7f32')
    plt.show()
```
