```python
filepath = '../../data/analysis/complete_data.csv'
```

```python
if not os.path.exists( filepath ):
    print("Missing dataset file")
else:
    df = pd.read_csv(filepath , encoding = "ISO-8859-1")
    print("File Read")
```

```python
df.columns
```

```python
df = df.groupby('Host_City')[['Year','Summer']].last().reset_index()
df
```

```python
folders = []
for i in df.index:
    year = df.loc[i].Year
    city = df.loc[i].Host_City
    summ = 'Winter'
    if df.loc[i].Summer:
        summ = 'Summer'
    if city == 'Montréal':
        city = 'Montreal'
    folders.append(str(year)+"_"+city+"_"+summ)
```

```python
path = '../../data/analysis/Games_Networks/{}/edges.csv'
paths = [path.format(i) for i in folders]
paths
```

```python
dfs =[]
for filepath in paths:
    if not os.path.exists( filepath ):
        print("Missing dataset file")
    else:
        try:
            dfs.append(pd.read_csv(filepath , encoding = "ISO-8859-1"))
            print("File Read")
        except:
            print("Error - "+filepath)
```
