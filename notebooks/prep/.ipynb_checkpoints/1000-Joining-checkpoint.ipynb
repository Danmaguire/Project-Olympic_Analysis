# Joins the Medal Data with the Country Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os.path
```

```python
# Ensure the file exists
if not os.path.exists( "../../data/prep/Games/Games-950.csv" ):
    print("Missing dataset file")
else:
    atheletesDF = pd.read_csv( "../../data/prep/Games/Games-950.csv" , encoding = "ISO-8859-1")
    print("File Read")
```

```python
atheletesDF.head()
```

```python
if not os.path.exists( "../../data/prep/Countries/countries_400.csv" ):
    print("Missing dataset file")
else:
    gamesDF = pd.read_csv( "../../data/prep/Countries/countries_400.csv")
    print("File Read")
```

```python
gamesDF.head()
```

Because we onlu have country data on the games up to 2016 we must remove the 2018 games

```python
atheletesDF[atheletesDF['Year'] != 2018]['Year'].unique()
```

```python
atheletesDF = atheletesDF[atheletesDF['Year'] != 2018]
atheletesDF.Year.unique()
```

Nationality of athletes is done via country code 

```python
atheletesDF['NOC'].unique()
```

```python
temp = atheletesDF[atheletesDF.NOC == 'EUN']
temps = temp[temp.Summer].index
tempw = temp[temp.Winter].index
for i in temps:
    atheletesDF.loc[i,'NOC'] = 'EUNS'
for i in tempw:
    atheletesDF.loc[i,'NOC'] = 'EUNW'
```

```python
for i in atheletesDF[atheletesDF.NOC == 'GDR'].index:
    atheletesDF.loc[i,'NOC'] = 'GERS'
```

```python
for i in atheletesDF[atheletesDF.NOC == 'FRG'].index:
    atheletesDF.loc[i,'NOC'] = 'GERS'
```

```python
new_df = pd.merge(atheletesDF, gamesDF,  how='left', left_on=['NOC','Year'], right_on = ['Code','Year'])
```

```python
new_df
```

```python
new_df.isnull().sum()
```

## Dataset for Education Analysis
Selects data relevant to the ananlysis of education as a factor effecting olympic success

Looks at what columns are in the dataset and which represent relevant data 

```python
new_df.columns
```

```python
edu = new_df[['Year', 'Host_Country', 'Host_City', 'Summer', 'Winter', 'Total_Males',
               'Total_Females', 'Total_Athletes', 'Discipline', 'Sport', 'Ath_Name',
               'Gender', 'Home_Adv', 'Gold', 'Silver', 'Bronze', 'Total_Medals',
               'Ath_Rating', 'Ath_Rank', 'NOC', 'NOC_Males_Sent', 'NOC_Females_Sent',
               'NOC_Total_Sent', 'NOC_Gold', 'NOC_Silver', 'NOC_Bronze',
               'NOC_Total_Medals', 'NOC_Rating', 'NOC_Rank', 'Country',
               'Education_Index']]
```

The education indexs are avaliable only for the years between 1980 and 2014 so we will limit this dataset to these years#


```python
edu = edu[new_df.Year <2014]
edu = edu[edu.Year >=1980]
ed = edu
edu.Year.unique()
```

```python
ed = len(edu[edu.Education_Index.isnull()])
tot = len(edu)
print(str(ed/tot*100)+"% Nulls")

edu.to_csv('../../data/analysis/education_data.csv')
```
