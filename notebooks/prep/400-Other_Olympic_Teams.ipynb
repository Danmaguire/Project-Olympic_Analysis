```python
codes = []
def getCode(country):
    try:
        code = dicDF.loc[country]['Code']
    except:
        if country == 'Bahamas, The':
            code = 'BAH'
        elif country == 'Congo, Dem. Rep.':
            code = 'COD'
        elif country == 'Egypt, Arab Rep.':
            code = 'EGY'
        elif country == 'Gambia, The':
            code = 'GAM'
        elif country == 'Korea, Dem. People?s Rep.':
            code = 'PRK'
        elif country == 'Micronesia, Fed. Sts.':
            code = 'FSM'
        elif country == 'Venezuela, RB':
            code = 'VEN'
        elif country == 'Virgin Islands (U.S.)':
            code = 'ISV'
    codes.append(code)
```

# addRows Function adds rows for the countries

```python
def addRows(year,country,code,region): 
    totPop = 0
    totM = 0
    totF = 0
    avgLE = 0
    totGDP = 0
    avgElevation = 0
    totArea = 0
    long = np.nan
    lat = np.nan
    popDensity = 0
    totco2 = 0
    totmeth = 0
    totnit = 0
    totEm = 0
    emPer = 0
    avgEdu = 0

    count=0
    cnt =0 
    temp = df[df.Year == year]
    temp = temp.set_index('Country')
    for member in members:
        row = temp.loc[member]
        totPop += row.Population
        totM += row.Males
        totF += row.Females
        avgLE += row.Life_Expectancy
        if not np.isnan(row.GDP):
            totGDP += row.GDP
        avgElevation += row.Elevation
        totArea += row.Area_SqKM
        if not np.isnan(row.CO2_Emissions):
            totco2 = row.CO2_Emissions
        if not np.isnan(row.Methane_Emissions):
            totmeth = row.Methane_Emissions
        if not np.isnan(row.Nitrous_Oxide_Emisions):
            totnit = row.Nitrous_Oxide_Emisions
        if not np.isnan(row.Education_Index):
            cnt +=1
            avgEdu += row.Education_Index


    avgElevation /= len(members)
    avgLE /= len(members)
    popDensity = totPop/totArea
    totEm = totco2+totmeth+totnit
    emPer = totEm/totPop
    if cnt == 0:
        avgEdu = np.nan
    else:
        avgEdu /= cnt
   
    max_elev = df[df.Country.isin(members)].Highest_Point.max()
    min_elev = df[df.Country.isin(members)].Lowest_Point.min()
    rangeel = max_elev-min_elev

    row = [country,year,totPop,totM,totF,avgLE,totGDP,region,avgElevation,totArea,long,lat,popDensity,totco2,totmeth,totnit,
       totEm,emPer,code,min_elev,max_elev,rangeel,avgEdu]
    
    df.loc[len(df)] = row
df.tail()
```

# Taiwan (TPE)<br>
The world bank unfortuantly do not keep data on Taiawan and therefore it is not contained within our dataframe<br>

```python
taiDF.Population = taiDF.Population*1000
```

```python
taiDF['Area'] = taiDF.Population/taiDF['Density (km²)']
taiDF['Area'] = taiDF.Area.mean()
taiDF.head()
```

```python
df.columns
```

```python
taiDF['Country'] = 'Taiwan'
taiDF = taiDF[['Country','Year','Population','% Male','% Female','Area']]
taiDF.head()
```

Years taiwan won medals in the games

```python
years = [1960,1968,1984,1992,1996,2000,2004,2008,2012,2016]
```

Geographic data on Taiwan

Years already in dataframe

```python
taiDF[taiDF.Year.isin(years)].Year.unique()
```

Years that still must be found must be found using estimates based on the nearest years currently in the dataframe

```python
remyears =  [1968,1984,1992,1996,2004,2008,2012]
prevYears = [1965,1980,1990,1995,2000,2005,2010]
nxtYears =  [1970,1985,1995,2000,2005,2010,2015] 
```

```python
for i in range(len(remyears)):
    year = remyears[i]
    prevY = prevYears[i]
    nxtY = nxtYears[i]
    prev = taiDF.set_index('Year').loc[prevY]
    nxt = taiDF.set_index('Year').loc[nxtY]


    pop = prev.Population + int(round(((nxt['Population']-prev.Population)/5)*(year-prevY)))
    maleP = (prev['% Male']+nxt['% Male'])/2
    female = 1-maleP
    popDen = pop/prev.Area
    taiDF.loc[len(taiDF)] = ['Taiwan',year,pop,maleP,female,prev.Area]
taiDF.tail()
```

```python
taiDF['Males'] = taiDF.Population*taiDF['% Male']
taiDF['Males'] = taiDF['Males'].apply(lambda x: int(round(x)))
taiDF['Females'] = taiDF.Population - taiDF.Males
taiDF['Life_Expectancy'] = np.nan
taiDF = taiDF[taiDF.Year.isin(years)]
taiDF.head()
```

```python
# Source - https://www.cia.gov/library/publications/the-world-factbook/fields/print_2020.html
mean_elevation = 1150
taiDF['Elevation'] = mean_elevation

low_point = 0
high_point = 3952
taiDF['Lowest_Point'] = low_point
taiDF['Highest_Point'] = high_point
taiDF['Elevation_Range'] =  high_point

# Source - https://www.indexmundi.com/factbook/fields/geographic-coordinates
long = 121
lat = 23.5
taiDF['Centroid_Longitude'] = long
taiDF['Centroid_Latitude'] = lat 
```
